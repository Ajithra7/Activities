```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("company_sales_data.csv")
data
```

```python
updated = data[(data["month"] == "FEBRUARY") | (data["month"] == "MARCH") | (data["month"] == "APRIL") | (data["month"] == "JULY") | (data["month"] == "AUGEST")]
updated
```

```python
col_filter = updated[["month", "facecream", "facewash"]]
col_filter
```

```python
plt.rcParams["figure.figsize"] = (6,4)
updated.plot(x = "month", y = ["facewash", "facecream"], kind = "bar",color = ["#cfcf57", "#9c1d23"])
plt.grid(color='gray', linestyle='dashed')
plt.xticks(rotation = 360)
plt.xlabel("month")
plt.ylabel("Sales units in Number")
plt.suptitle("Facewash Sales Vs Facecream Sales \n", fontweight="bold")
plt.title("For the month facewash sales less than 1500")
plt.show()
```
